```python
df['radius_mean'] >= 13.37 # aplicara um teste condicional para a coluna radius
```

```python
# Vamos criar uma nova coluna apartir dos resultados da condicional
df.loc[(df['radius_mean'] >= 13.37),'class_radius'] = 'H' # A coluna vai receber um nome e recebara para cada valor que passar 
```

```python
df.loc[(df['radius_mean'] < 13.37),'class_radius'] = 'L'
```

```python
df.drop('Unnamed: 32', axis=1, inplace=True)
```

```python
df.drop('class_radius', axis=1, inplace=True)
```

```python
import numpy as np
```

```python
# Colocamos as condicoes em listas e as escolhas o que sera colocado caso a condicional seja aceita
# na outra lista na ordem respectiva.
cond = [df['radius_mean'] >= 13.37, df['radius_mean'] < 13.37]
esc = ['H', 'L']
```

```python
# Select =Return an array drawn from elements in choicelist, depending on conditions.
df['class_radius'] = np.select(cond, esc, default=None)
```

```python
# Agrupa os dados pelo High ou Low e calcula a media(ou qualquer outro parametros) do agrupamento
df.groupby('class_radius').mean()
```

```python
# Agrupa os dados pelo high ou low e tbm pelo Maligno e benigno e calcula um parametros
df.groupby(['diagnosis','class_radius']).mean()
```

```python
ID = df['id'].sort_values().tolist() # Vai ordenar e vai colocar os valores em uma lista
# guardando o resultado na variavel ID
```

```python
df['id']
```

```python
df2 = pd.DataFrame(data={'id':ID, 'sex':sex, 'age':age}) # Criando o segundo Dataframe
# data frame, com os o valores chave:valor(lista)
```

```python
df2
```

```python
# Unindo os dois Df com a funcao merge, o parametro da segundo df fica na direita e uni atraves do id
"""Esse código usa a biblioteca Pandas para unir dois DataFrames (df e df2) usando a função merge().

A função merge() é usada para combinar dois DataFrames baseados em uma coluna (neste caso, a coluna "id"). A opção "how='inner'" especifica que a união deve ser um "inner join", o que significa que apenas as linhas que têm valores correspondentes em ambos os DataFrames serão incluídas no resultado.

O resultado da união é armazenado em uma nova variável chamada "dfMerge"."""
dfMerge = df.merge(df2, how='inner',on = 'id') # intersecao,
```

```python
df3 = dfMerge.iloc[:10,:]*0
```

```python
df3
```

```python
pd.concat([dfMerge,df3], ignore_index=True) # COncatena dois df no final
# ignore_index ignora o indice
```

```python
dfMerge.pivot(index='id', columns='diagnosis', values='radius_mean').fillna(0)
# fillna prenche os dados em branco com um valor
"""Esse código usa a função pivot() do Pandas para transformar os dados de forma longa para forma larga. A função pivot() agrupa os valores em uma tabela pivotada (também conhecida como tabela dinâmica) onde uma coluna é transformada em várias colunas.

A opção "index='id'" especifica que a coluna "id" deve ser usada como índice (linhas) na tabela pivotada. A opção "columns='diagnosis'" especifica que a coluna "diagnosis" deve ser usada como colunas na tabela pivotada. E a opção "values='radius_mean'" especifica que os valores na tabela pivotada devem ser preenchidos com os valores da coluna "radius_mean".

A função fillna(0) é usada para preencher quaisquer valores NaN (Not a Number) na tabela pivotada com zero.

O resultado da transformação é armazenado na mesma variável "dfMerge", substituindo seus dados anteriores."""
```

```python
x = dfMerge.pivot_table(index='sex', columns='diagnosis', values='radius_mean', aggfunc='count').T
# aggfunc pode receber, std, mean, median, max,count, etc.
"""Esse código usa a função pivot_table() do Pandas para transformar os dados de forma longa para forma larga e agrupar os dados ao mesmo tempo. A função pivot_table() é semelhante à função pivot(), mas fornece mais recursos de agrupamento de dados.

A opção "index='sex'" especifica que a coluna "sex" deve ser usada como índice (linhas) na tabela pivotada. A opção "columns='diagnosis'" especifica que a coluna "diagnosis" deve ser usada como colunas na tabela pivotada. A opção "values='radius_mean'" especifica que os valores na tabela pivotada devem ser preenchidos com os valores da coluna "radius_mean". E a opção "aggfunc='count'" especifica que a função de agregação a ser usada é a função de contagem "count".

O resultado da transformação e agrupamento é armazenado na mesma variável "dfMerge", substituindo seus dados anteriores.

A tabela pivotada resultante irá exibir a contagem de valores da coluna "radius_mean" agrupados por valores na coluna "sex" e colunas "diagnosis"."""
```
